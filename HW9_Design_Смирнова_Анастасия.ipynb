### Вопрос 1
Ответ: Батч-архитектура с обучением по расписанию
Обоснование: Наиболее оптимальная архитектура для данной задачи - батчевая. Потоковая и онлайн архитектуры слишком избыточно и создадут дополнительную нагрузку, поскольку подразумевают частые обновления, чего не требуется в данной задаче. Гибридная архитектура допустима, но также является избыточной, поскольку в нашей задаче не требуется многновенных обновлений. Переобучение модели лучше запускать по расписанию, поскольку требуется обновление раз в сутки. Для этого отлично подойдут CRON задачи. Обновление по появлению новых данных избыточно, так как при появлении каждого нового батча, будет генерироваться нагрузка на систему. Остальные два варианта также не удовлетворяют требованию "ежедневное обновление"

### Вопрос 2
Ответ: Потоковая архитектура с запуском переобучения при появлении новых данных
Обоснование: Поскольку критически важно обновление и предсказания в реальном времени, наиболее оптимальным варинатом будет потоковая архитектура. Остальные типы архитектуры не смогут обеспечить нужной оперативности. Обновление необходимо запускать по появлению новых данных, чтобы модель всегда имела представление о наиболее свежих данных.

### Вопрос 3
Ответ: Гибридная архитектура в сочетании с переобучением по алерту мониторинга
Обоснование: Для данной задачи лучше всего подойдет гибридная архитектура с запуском переобучения по алерту мониторинга (например, при обнаружении data drift или concept дрифт). Такой подход позволит моментально учесть изменения в поведении пользователей в сочетании с предсказаниями на больших объемах данных.

### Вопрос 4
Ответ: Онлайн-обучение с обновлением при появлении новых данных
Обоснование: Поскольку для данной задачи важна постоянная адаптация к меняющимся условиям больше всего подойдет онлайн-обучение в сочетании с обновлением по появлению новых данных. Это напрямую соответствует необходимости инкрементального обновления.

```python
%%writefile generate_synthetic_data.py
# Вспомогательный скрипт для искусственной генерации данных
import boto3
from botocore.client import Config
import os

ENDPOINT_URL = os.getenv("S3_ENDPOINT", "http://localhost:4566")
BUCKET = "retail-checks"
KEY = "processed/total_checks_sum.txt"

def get_s3():
    return boto3.client(
        "s3",
        endpoint_url=ENDPOINT_URL,
        aws_access_key_id="fake",
        aws_secret_access_key="fake",
        config=Config(signature_version="s3v4"),
        region_name="us-east-1"
    )

def set_checks(count: int):
    s3 = get_s3()
    try:
        s3.head_bucket(Bucket=BUCKET)
    except:
        s3.create_bucket(Bucket=BUCKET)
    s3.put_object(Bucket=BUCKET, Key=KEY, Body=str(count).encode())
    print(f"Записано чеков: {count}")

if __name__ == "__main__":
    import argparse
    parser = argparse.ArgumentParser()
    parser.add_argument("--checks", type=int, required=True)
    args = parser.parse_args()
    set_checks(args.checks)
```

Для решения нашей задачи создадим два DAG: `ml_inventory_retraining_pipeline.py` и `ml_inventory_monitor.py`.

Первый DAG отвечает за обучение. 
Второй DAG проверяет условия и в случае срабатывания, запускает пайплайн.

```python
%%writefile dags/ml_inventory_retraining_pipeline.py
from airflow import DAG
from airflow.operators.python import PythonOperator
from datetime import datetime
import random

def train_model(**context):
    print("Загрузка чеков с S3... обучение линейной регрессии...")
    print("Модель обучена.")

def evaluate_model(**context):
    accuracy = random.uniform(0.8, 0.95)
    print(f"Модель оценена. Accuracy = {accuracy:.2f}")
    if accuracy < 0.85:
        raise ValueError("Качество модели ниже порога, деплой отменён.")
    context['ti'].xcom_push(key='new_accuracy', value=accuracy)

def deploy_model(**context):
    print("Модель задеплоена.")

with DAG(
    dag_id="ml_inventory_retraining_pipeline",
    start_date=datetime(2025, 1, 1),
    schedule=None,
    catchup=False,
    max_active_runs=1,
    description="Continuous Training pipeline для расчёта складских запасов",
    tags=["ml", "inventory"]
) as dag:
    train = PythonOperator(task_id="train_model", python_callable=train_model)
    evaluate = PythonOperator(task_id="evaluate_model", python_callable=evaluate_model)
    deploy = PythonOperator(task_id="deploy_model", python_callable=deploy_model)
    train >> evaluate >> deploy
```

```python
%%writefile dags/ml_inventory_monitor.py
from airflow import DAG
from airflow.operators.python import PythonOperator, BranchPythonOperator
from airflow.operators.empty import EmptyOperator
from airflow.operators.trigger_dagrun import TriggerDagRunOperator
from airflow.providers.amazon.aws.hooks.s3 import S3Hook
from airflow.models import Variable
from datetime import datetime, timedelta

def check_conditions(**context):
    need_retrain = False
    # S3-триггер
    try:
        s3 = S3Hook(aws_conn_id="aws_default")
        if s3.check_for_key("processed/total_checks_sum.txt", bucket_name="retail-checks"):
            obj = s3.get_key("processed/total_checks_sum.txt", bucket_name="retail-checks")
            total_checks = int(obj.get()['Body'].read().decode().strip())
            if total_checks > 10_000_000:
                print(f"S3 trigger: чеков {total_checks} (>10M)")
                need_retrain = True
    except Exception as e:
        print(f"S3 check failed: {e}")

    # Accuracy-триггер
    current_accuracy = float(Variable.get("current_accuracy", default_var="0.95"))
    if current_accuracy < 0.85:
        print(f"Accuracy trigger: {current_accuracy} < 0.85")
        need_retrain = True

    # 3. Срок модели (через конфиг запуска)
    model_expiry_str = context['dag_run'].conf.get('model_expiry') if context.get('dag_run').conf else None
    if model_expiry_str:
        model_expiry = datetime.fromisoformat(model_expiry_str)   # преобразуем строку в datetime
    else:
        model_expiry = datetime.now() + timedelta(hours=2)

    if datetime.now() + timedelta(hours=1) >= model_expiry:
        print(f"Expiry trigger: expires at {model_expiry}")
        need_retrain = True

    context['ti'].xcom_push(key='need_retrain', value=need_retrain)

def branch_decision(**context):
    need_retrain = context['ti'].xcom_pull(key='need_retrain', task_ids='check_conditions')
    return 'trigger_retraining' if need_retrain else 'skip'

with DAG(
    dag_id="ml_inventory_monitor",
    start_date=datetime(2025, 1, 1),
    schedule="@hourly",
    catchup=False,
    max_active_runs=1,
    default_args={'owner': 'ml_team'},
    description="Монитор условий и запуск переобучения",
    tags=["ml", "monitor"]
) as dag:
    check = PythonOperator(task_id="check_conditions", python_callable=check_conditions)
    decide = BranchPythonOperator(task_id="branch", python_callable=branch_decision)

    trigger = TriggerDagRunOperator(
        task_id="trigger_retraining",
        trigger_dag_id="ml_inventory_retraining_pipeline",
        wait_for_completion=False
    )
    skip = EmptyOperator(task_id="skip")
    check >> decide >> [trigger, skip]
```

*Ожидаемый артефакт: код создания/деинсталляции инфраструктуры в [ячейке](#scrollTo=vhSxlazCtGF2)*


Создадим файл `ci-cd.yaml` для GitHub Actions.
Пайплайн содержит три джобы:

1) `lint-and-test` отвечает за проверку DAG.
Цель этой джобы убедиться в том, что пайплайн Airflow не содержит ошибок.

2) `terraform-check` отвечает за валидацию terraform кода.
Цель этой джобы убедиться в том, что `.tf-файлы` синтаксически верны и `Terraform` сможет их интерпретировать.

3) `infrastructure-demo` отвечает за демонстрацию полного цикла IaC от создания инфраструктуры до её гарантированного уничтожения.
`MinIO` выступает в качестве временного S3-совместимого сервера. `Terraform`, используя провайдер `AWS`, обращается к этому `MinIO` (endpoint http://localhost:9000) и создаёт бакеты `retail-checks-demo` и `ml-model-registry-demo`.

После проверки (`aws s3 ls`) `Terraform destroy` удаляет эти бакеты.


Также для демонстрации было создано три `.tf файла`:
- `main.tf` определяет, какие бакеты создавать , и настраивает провайдер под `MinIO`.
- `variables.tf` позволяет параметризовать окружение (`environment=demo`).
- `outputs.tf` выводит имена созданных бакетов (можно увидеть в плане).

[Пайплайн успешно выполнен](https://github.com/steishas/hw-9-mlops-mipt-smirnova-anastasia/actions/runs/26330786768).

Таким образом, инфраструктура не создается навсегда: она создаётся, проверяется и сразу уничтожается. Это подтверждает, что код поддерживает деинсталляцию ресурсов и не требует ручной очистки. 

```python
%%writefile .github/workflows/ci-cd.yaml
name: ML Continuous Training CI/CD

on:
  push:
    branches: [ main ]
  pull_request:
    branches: [ main ]

env:
  DAGS_FOLDER: ./dags

jobs:
  # 1. Линтинг DAG и юнит-тесты
  lint-and-test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - name: Set up Python
        uses: actions/setup-python@v5
        with:
          python-version: "3.11"
      - name: Install dependencies
        run: pip install -r requirements.txt
      - name: Lint DAGs
        run: python -m py_compile ${{ env.DAGS_FOLDER }}/*.py
      - name: Unit tests
        run: PYTHONPATH=. pytest tests/

  # 2. Валидация Terraform-кода
  terraform-check:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: hashicorp/setup-terraform@v3
        with:
          terraform_version: "1.8.0"
      - name: Terraform Init
        run: terraform init
      - name: Terraform Validate
        run: terraform validate

  # 3. Полная демонстрация IaC (создание -> проверка -> удаление)
  infrastructure-demo:
    needs: terraform-check
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: hashicorp/setup-terraform@v3
        with:
          terraform_version: "1.8.0"

      - name: Start MinIO
        run: |
          docker run -d --name minio \
            -p 9000:9000 -p 9001:9001 \
            -e MINIO_ROOT_USER=minioadmin \
            -e MINIO_ROOT_PASSWORD=minioadmin \
            quay.io/minio/minio server /data --console-address ":9001"

      - name: Wait for MinIO
        run: |
          for i in $(seq 1 30); do
            if curl -s http://localhost:9000/minio/health/live; then
              echo "MinIO готов"
              break
            fi
            echo "Ожидание MinIO..."
            sleep 2
          done

      - name: Terraform Init
        run: terraform init

      - name: Terraform Plan
        env:
          AWS_ACCESS_KEY_ID: minioadmin
          AWS_SECRET_ACCESS_KEY: minioadmin
        run: terraform plan -var="environment=demo"

      - name: Terraform Apply
        env:
          AWS_ACCESS_KEY_ID: minioadmin
          AWS_SECRET_ACCESS_KEY: minioadmin
        run: terraform apply -auto-approve -var="environment=demo"

      - name: Check created buckets
        env:
          AWS_ACCESS_KEY_ID: minioadmin
          AWS_SECRET_ACCESS_KEY: minioadmin
        run: aws --endpoint-url=http://localhost:9000 s3 ls

      - name: Terraform Destroy
        if: always()
        env:
          AWS_ACCESS_KEY_ID: minioadmin
          AWS_SECRET_ACCESS_KEY: minioadmin
        run: terraform destroy -auto-approve -var="environment=demo"

      - name: Stop MinIO
        if: always()
        run: docker rm -f minio
```

### Управление рисками: SLI/SLO на трёх уровнях

#### Бизнес-уровень
- SLI: Доля заказов с задержкой отгрузки >24 часов по причине «нехватка товара на складе» (фиксируется менеджерами в системе). Измеряется помесячно.
- SLO: <= 5% заказов за месяц.
- Реакция: При превышении порога инициируется внеплановая проверка модели и складских политик с участием ответственного менеджера.

#### Уровень модели и данных (код)
- SLI-1 (качество прогнозов): `MAPE` прогноза списания на отложенной 7-дневной выборке.
- SLO-1: `MAPE` <= 15%. При 15-20% — алерт ML-инженерy. При >20% — (аварийное переобучение + ручная валидация).
---
- SLI-2 (data drift): `Population Stability Index (PSI)` ключевых признаков (средний чек, частота покупок), рассчитанный между последним днём и эталонным распределением за предыдущий месяц.
- SLO-2: PSI ≤ 0.2. При превышении — алерт, запуск углублённого анализа дрейфа. Мониторинг реализован через дашборд Evidently, который периодически обновляется задачей Airflow.
---
- SLI-3 (своевременность обучения): Время между появлением нового батча чеков в `S3` и успешным завершением`DAG` переобучения.
- SLO-3: <= 25 часов (запас после ежедневного расписания). Если >30 часов — уведомление дежурного инженера.

#### Инфраструктурный уровень (железо)
- SLI-1 (доступность объектного хранилища): Процент успешных S3-запросов (`MinIO`) за 5-минутные интервалы.
- SLO-1: <= 99.9% в месяц. При падении ниже 99% — критический алерт.
---
- SLI-2 (успешность пайплайнов): Доля успешных запусков DAG ml_inventory_retraining_pipeline за неделю.
- SLO-2: 100% (допускается не более 1 сбоя в месяц). Два подряд неудачных запуска — вмешательство инженера.

#### Каналы оповещения:
 Все алерты отправляются на e-mail ответственной команды (ML-инженер, менеджер по поставкам) с деталями сработавшего индикатора и ссылкой на соответствующий дашборд.

#### Взаимосвязь уровней:
 Деградация бизнес-показателя (задержки отгрузок) в первую очередь анализируется через SLI модели и инфраструктуры. Например, рост задержек может быть вызван высокой ошибкой прогноза (`MAPE`), дрейфом данных или пропуском переобучения из-за сбоя `S3/DAG`. Для данной задачи возьмем бизнес SLI: долю заказов с задержкой отгрузки более 24 часов по причине "нехватка товара на складе". 
Так как данная метрика является долевой, для оценки ее изменений необходимо использовать z-тест пропорций.

Создадим две выборки:
- 1000 заказов и 5% задержанных;
- 500 заказов и 7% задержанных.

Другие вводные данные для эксперимента:
- $\alpha$ - уровень значимости 0.05;
- $\beta$ - доверительный интервал 0.95.
- $H_0$: Ухудшения метрики не зафиксировано;
- $H_1$: Зафиксировано статистически значимое ухудшение метрики.

Для проведения теста будем использовать `proportions_ztest` из `statsmodels.stats.proportion`.


```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Исторические данные (референс)
n_hist = 1000
x_hist = 50   # 5%
p_hist = x_hist / n_hist

# Новые данные
n_new = 500
x_new = 35    # 7%
p_new = x_new / n_new

# z-тест (односторонний: новый > исторический)
count = np.array([x_new, x_hist])
nobs = np.array([n_new, n_hist])
stat, p_value = proportions_ztest(count, nobs, alternative='larger')

# Доверительный интервал для разности пропорций (новый - исторический)
diff = p_new - p_hist
se = np.sqrt(p_hist*(1-p_hist)/n_hist + p_new*(1-p_new)/n_new)
ci_diff_low = diff - 1.96 * se
ci_diff_upp = diff + 1.96 * se

print(f"Историческая доля: {p_hist:.3f}")
print(f"Новая доля: {p_new:.3f}")
print(f"Разность: {diff:.3f}")
print(f"95% CI для разности: [{ci_diff_low:.3f}, {ci_diff_upp:.3f}]")
print(f"z-статистика: {stat:.3f}, p-value (односторонний): {p_value:.3f}")

if p_value > 0.05:
    print("Нет достаточных оснований отвергнуть нулевую гипотезу: изменение незначимо.")
    print("Решение: ничего не менять, продолжить мониторинг.")
else:
    print("Отвергаем нулевую гипотезу: ухудшение статистически значимо.")
    print("Решение: запустить внеплановое улучшение модели/процесса.")
```

p_value > alpha, поэтому у нас не оснований отвергать нулевую гипотезу.
95% интервал для разности включает 0, что с вероятностью 0.95 указывает на то, что в реальности доля задержек могла и не вырасти.
На основании полученных результатов нет оснований вносить изменения, необходимо продолжить мониторинг.

Составим ADR.

```python
%%writefile docs/adr-001.md
# Отказ от внепланового переобучения модели при незначимом росте доли задержек

Дата: 2026-05-23
Статус: предложено

## Контекст
Бизнес-SLI «доля заказов с задержкой по причине нехватки товара» за исторический месяц (1000 заказов) составляет 5%. За последнюю неделю (500 заказов) эта доля выросла до 7%. SLO — <= 5%. Необходимо решить, требуется ли внеплановое улучшение ML-системы (переобучение, пересмотр страховых запасов).

## Решение
Мы провели односторонний z-тест для двух независимых пропорций. Получено p-value = 0.057, что выше порога значимости 0.05. 95% доверительный интервал для разности долей (новый – исторический) составил [–0.006, 0.046] и включает ноль. Следовательно, наблюдаемый рост не является статистически значимым. Мы принимаем решение не запускать внеплановое переобучение модели и не изменять текущие параметры системы.

## Последствия
- Риск: реальное ухудшение может проявиться с накоплением данных и привести к нарушению SLO (>5% в месяц).
- Смягчение: установить дополнительный еженедельный мониторинг и повторить тест через неделю с увеличенной выборкой.
- Выгода: экономим ресурсы ML-инженера и вычислительные мощности, избегая необоснованных изменений.
- Ответственный: ML-инженер оповещается автоматически при следующем подобном превышении, решение о вмешательстве принимается коллегиально.
```
